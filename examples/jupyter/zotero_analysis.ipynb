# Zotero Data Analysis

This notebook demonstrates how to use the `zotero-client` library to interact with your Zotero library, fetch data, and perform some basic analysis.

```
# 1. Configure and Initialize the Zotero Client
# Make sure your .env file is set up with your ZOTERO_API_KEY and ZOTERO_USER_ID
from dotenv import load_dotenv
import os
from zotero_client.api.client import ZoteroClient

load_dotenv()

api_key = os.getenv('ZOTERO_API_KEY')
user_id = os.getenv('ZOTERO_USER_ID')

client = ZoteroClient(api_key=api_key, user_id=user_id)
```

```
# 2. Fetch Items from Your Library
# This example fetches the 10 most recently modified items.
items = client.get_items(limit=10)

for item in items:
    print(f"- [{item.item_type}] {item.title}")
```
